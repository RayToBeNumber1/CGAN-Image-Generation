```python
def generate_one_image(generator, test_image_path):
    input_image = tf.io.read_file(test_image_path)
    input_image = tf.io.decode_jpeg(input_image, channels=3)
    input_image = tf.cast(input_image, tf.float32)
    input_image = tf.image.resize(input_image, [256, 256],
                                method = tf.image.ResizeMethod.NEAREST_NEIGHBOR)
    input_image = (input_image / 127.5) - 1
    input_image = tf.reshape(input_image, [1,256,256,3])
    image = generator(input_image, training=True)
    image = tf.reshape(image, [256,256,3])
    plt.imshow(image*0.5+0.5)
```

```python
test_image_dir_path = '/content/drive/MyDrive/CGAN_PIX2PIX/train_dataset/input'
```

```python
def generate_many_images(generator, test_image_dir_path):
    plt.subplots(100,1,figsize = (256,256))

    for i in range(1,101):
        input_image = tf.io.read_file(f'{test_image_dir_path}/{i}.png')
        input_image = tf.io.decode_jpeg(input_image, channels=3)
        input_image = tf.cast(input_image, tf.float32)
        input_image = tf.image.resize(input_image, [256, 256],
                                    method = tf.image.ResizeMethod.NEAREST_NEIGHBOR)
        input_image = (input_image / 127.5) - 1
        input_image = tf.reshape(input_image, [1,256,256,3])
        image = generator(input_image, training=True)
        image = tf.reshape(image, [256,256,3])
        plt.subplot(100,1,i)
        plt.imshow(image*0.5+0.5)
```

```python
generate_many_images(generator)
```
